```python
def progression(consonants, liquids, glides, nasals, voiceless_obs, voiced_obs, LSNInt, LSNFin, LSN):
    
    cons=consonants
    articulation=[liquids, nasals, glides, voiceless_obs, voiced_obs]
    Amb=set()
    LevelOne=set()
    LevelTwo=set()
    LevelThree=set()
    LevelFour=set()
    LevelFive=set()
    LevelSix=set()
    LevelSeven=set()
    LevelEight=set()
    LevelNine=set()
    LevelTen=set()
    Other=set()
    
    levels=[
            LevelOne, 
            LevelTwo,
            LevelThree,
            LevelFour,
            LevelFive,
            LevelSix,
            LevelSeven, 
            LevelEight,
            LevelNine, 
            LevelTen
    ]
    
    
    for (g,p) in consonants.items():
        if len(p)>1: 
            Amb.add((g,p))        
    
    for gp in consonants:
        
        if gp in articulation[0]: 
            if gp in LSN:
                if gp in LSNInt:
                    LevelOne.add((gp, 'LSN_I'))
                elif gp in LSNFin: 
                    LevelOne.add((gp, 'LSN_F'))
            else: 
                Other.add(gp)
        
        elif gp in articulation[1]:
            if gp in LSN:
                if gp in LSNInt:
                    LevelTwo.add((gp, 'LSN_I'))
                elif gp in LSNFin: 
                    LevelTwo.add((gp, 'LSN_F'))
            else: 
                Other.add(gp)
        
        elif gp in articulation[2]:
            if gp in LSN:
                if gp in LSNInt:
                    LevelThree.add((gp, 'LSN_I'))
                elif gp in LSNFin: 
                    LevelThree.add((gp, 'LSN_F'))
            else: 
                Other.add(gp)
        
        elif gp in articulation[3]:
            if gp in LSN:
                if gp in LSNInt:
                    LevelFour.add((gp, 'LSN_I'))
                elif gp in LSNFin: 
                    LevelFour.add((gp, 'LSN_F'))
            else: 
                Other.add(gp)
        
        else:
            if gp in articulation[4]:
                if gp in LSN:
                    if gp in LSNInt:
                        LevelFive.add((gp, 'LSN_I'))
                    elif gp in LSNFin: 
                        LevelFive.add((gp, 'LSN_F'))
                else: 
                    Other.add(gp)

    for gp in Other: 
        if gp in articulation[0]:
            LevelSix.add(gp)
        elif gp in articulation[1]:
            LevelSeven.add(gp)
        elif gp in articulation[2]:
            LevelEight.add(gp)
        elif gp in articulation[3]:
            LevelNine.add(gp)
        elif gp in articulation[4]:
            LevelTen.add(gp)
        
        
    LevelNo=1
    for item in levels: 
        if len(item)==0: 
            continue
        else:
            print(f'Level {LevelNo}: {item}')
            LevelNo+=1
    
    
                
        
    
    
    
    
```

```python
progression(consonants,liquids, glides, nasals, voiceless_obs, voiced_obs, LSNIn, LSNFin, LSN) 
```
